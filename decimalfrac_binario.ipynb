### Conversor de Decimal Fracionário para Binario Fracionario ( e o oposto também )

Tive um pouco de dificuldade nesse aqui, talvez o código não tenha ficado o mais limpo possível, mas o problema foi resolvido. Acabei simplificando e utilizando o que eu conhecia para resolver.

```python
variavelSentinela = True

while variavelSentinela == True:

    operacao = int(input("1: Decimal->Binario\n2: Binario->Decimal\n0: Sair\n"))

    if operacao == 1:
        numero_decimal = float(input("DECIMAL FRACIONARIO: "))
        parte_inteira = int(numero_decimal)
        parte_frac = numero_decimal - parte_inteira
        lista_inteira = []
        lista_frac = []

        while parte_inteira != 0:
            resto = parte_inteira % 2
            lista_inteira.append(resto)
            parte_inteira = parte_inteira // 2

        lista_inteira.reverse()

        while parte_frac % 1 != 0:
            n = int(parte_frac * 2)
            lista_frac.append(n)
            parte_frac = parte_frac*2 - n

        numero_inteiro_binario = ""
        for numero in lista_inteira:
            numero_inteiro_binario += str(numero)

        numero_frac_binario = ""
        for numero in lista_frac:
            numero_frac_binario += str(numero)


        print(f"{numero_inteiro_binario},{numero_frac_binario}\n")

    elif operacao == 2:
        numero_binario = input("BINARIO FRACIONARIO: ")
        parte = numero_binario.split(",")

        parte_inteira = parte[0]
        parte_frac = parte[1]

        expoente = len(parte_inteira) - 1
        numero_decimal = 0

        for n in parte_inteira:
            numero_decimal += int(n) * 2**expoente
            expoente -= 1

        for m in parte_frac:
            numero_decimal += float(m) * 2**expoente
            expoente -= 1

        print(f"{numero_decimal}\n")
    elif operacao == 0:
        variavelSentinela = False
    else:
        print("Inválido!")
```
