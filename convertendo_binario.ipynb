### Convertendo Decimal em Binário

Fiz utilizando o metodo do resto pois me senti mais confortável

```python
numero_decimal = int(input())
lista_binario = []
numero_binario = ""

while numero_decimal != 0:
    resto = numero_decimal % 2
    lista_binario.append(resto)
    numero_decimal = numero_decimal // 2

lista_binario.reverse # usei pra inverter a lista, vi que você utilizou em um exemplo da aula e entendi que era permitido usar.

for numero in lista_binario:
    numero_binario += str(numero)

print(numero_binario)
```
